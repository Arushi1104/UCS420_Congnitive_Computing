```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Part I: Randomized Sales Data Generation

np.random.seed(7147)  # Use roll number as seed

# Generate sales data for 12 months and 4 product categories
sales_data = np.random.randint(1000, 5000, size=(12, 4))

# Define column names and month labels
columns = ["Electronics", "Clothing", "Home & Kitchen", "Sports"]
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]

# Convert to DataFrame
df = pd.DataFrame(sales_data, columns=columns, index=months)

# Display first 5 rows
print(df.head())

# Part II: Data Manipulation and Analysis

print(df.describe())  # Summary statistics

# Total sales per category
print(df.sum())

# Total sales per month
df["Total Sales"] = df.sum(axis=1)
print(df["Total Sales"])

# Calculate average sales growth per category
df_growth = df.iloc[:, :-1].pct_change() * 100
print(df_growth.mean())

# Add Growth Rate column
df["Growth Rate"] = df["Total Sales"].pct_change() * 100

# Apply discount based on roll number (Odd → 15% off Clothing, Even → 10% off Electronics)
if 7147 % 2 == 0:
    df["Electronics"] *= 0.90
else:
    df["Clothing"] *= 0.85

# Part III: Visualizations

# Line plot for monthly sales trends
plt.figure(figsize=(12, 6))
for col in df.columns[:-2]:
    plt.plot(df.index, df[col],  label=col)

plt.title("Monthly Sales Trends")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.legend()
plt.grid(True)
plt.show()

# Box plot for sales distribution by category
plt.figure(figsize=(10, 6))
sns.boxplot(data=df.iloc[:, :-2])
plt.title("Sales Distribution by Category")
plt.show()
```

```python
import numpy as np

# Given array
arr = np.array([[1, -2, 3], [-4, 5, -6]])

# i. Absolute values
abs_arr = np.abs(arr)
print("Absolute values:\n", abs_arr)

# ii. Percentiles (Flattened, Column-wise, Row-wise)
p_flat = np.percentile(arr, [25, 50, 75])
p_col = np.percentile(arr, [25, 50, 75], axis=0)
p_row = np.percentile(arr, [25, 50, 75], axis=1)

print("Percentiles (Flattened):", p_flat)
print("Percentiles (Columns):\n", p_col)
print("Percentiles (Rows):\n", p_row)

# iii. Mean, Median, Std Dev (Flattened, Column-wise, Row-wise)
m_flat, med_flat, s_flat = np.mean(arr), np.median(arr), np.std(arr)
m_col, med_col, s_col = np.mean(arr, axis=0), np.median(arr, axis=0), np.std(arr, axis=0)
m_row, med_row, s_row = np.mean(arr, axis=1), np.median(arr, axis=1), np.std(arr, axis=1)

print("Mean (Flattened):", m_flat, "| Median:", med_flat, "| Std Dev:", s_flat)
print("Mean (Columns):", m_col, "| Median:", med_col, "| Std Dev:", s_col)
print("Mean (Rows):", m_row, "| Median:", med_row, "| Std Dev:", s_row)
```

```python
import numpy as np

a = np.array([-1.8, -1.6, -0.5, 0.5, 1.6, 1.8, 3.0])

floor_values = np.floor(a)
ceiling_values = np.ceil(a)
truncated_values = np.trunc(a)
rounded_values = np.round(a)

print("Floor values:", floor_values)
print("Ceiling values:", ceiling_values)
print("Truncated values:", truncated_values)
print("Rounded values:", rounded_values)
```
